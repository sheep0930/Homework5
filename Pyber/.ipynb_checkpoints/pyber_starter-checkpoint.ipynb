```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
# Read the City and Ride Data
print (city_data_df.head())
print (ride_data_df.head())
ride_data_df.head()
# Combine the data into a single dataset
combined_df = pd.merge(ride_data_df, city_data_df, on="city", how="left")

combined_df.head()


# Display the data table for preview
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
combined_group = combined_df.groupby(['type','city'])
tot_ride_num = combined_group['city'].count()
ave_fare = combined_group['fare'].mean()
ave_fare
combined_type_df = pd.DataFrame()
combined_type_df['ride_num'] = tot_ride_num
# combined_type_df.head()
combined_type_df['ave_fare'] = ave_fare
# combined_type_df.head()
combined_type_df['driver_count'] = combined_group['driver_count'].mean()

print (combined_type_df.head())
rural_df = combined_type_df.loc['Rural']
urban_df = combined_type_df.loc['Urban']
suburban_df = combined_type_df.loc['Suburban']
print (suburban_df.head())

# Build the scatter plots for each city types
rural_handle = plt.scatter(rural_df['ride_num'], rural_df['ave_fare'],facecolors='gold', edgecolors="black",
            s=rural_df['driver_count']*10, alpha=0.75)
urban_handle = plt.scatter(urban_df['ride_num'], urban_df['ave_fare'],facecolors='lightcoral', edgecolors="black",
            s=urban_df['driver_count']*10, alpha=0.75)
suburban_handle = plt.scatter(suburban_df['ride_num'], suburban_df['ave_fare'],facecolors='lightskyblue', edgecolors="black",
            s=suburban_df['driver_count']*10, alpha=0.75)
# Incorporate the other graph properties
plt.grid()
plt.tight_layout()
plt.legend([urban_handle, suburban_handle,rural_handle],["Urban","Suburban","Rural"],title='City Types',loc="best")
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares ($)")
plt.text(42, 36, "Note:\nCircle size correlates with driver count per city.", fontsize=10)
plt.savefig("Pyber_Ride_Sharing_Data_2016.jpg")
# Create a legend

# Incorporate a text label regarding circle size


# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_group = combined_df.groupby('type')
tot_fare_df = type_group['fare'].sum()
labels = ["Rural", "Suburban", "Urban"]
# # Build Pie Chart
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)
plt.pie(tot_fare_df, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig("Total_Fares_by_City_Type.jpg")
```

```python
# Show Figure
plt.show()
```
